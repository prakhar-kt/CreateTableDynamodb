```python
# Get the list of repos after the repo id given 
def list_repos(token, since='512070020'):
    res = requests.get(
        f'https://api.github.com/repositories?since={since}',
        headers = {'Authorization': f'token {token}'}
        
    )
    return json.loads(res.content.decode('utf-8'))
```

```python
# Get repo details for a given owner name and token
def get_repo_details(owner, name, token):
    
    res = requests.get(
        f'https://api.github.com/repos/{owner}/{name}',
        headers = {'Authorization': f'token {token}'}
    )
    return json.loads(res.content.decode('utf-8'))
```

```python
def extract_repo_fields(repo):
    repo_fields = {
        'id': repo['id'],
        'node_id': repo['node_id'],
        'name': repo['name'],
        'owner': {
            'login': repo['owner']['login'],
            'id': repo['owner']['id'],
            'type': repo['owner']['type'],
            'site_admin': repo['owner']['site_admin']
        },
        'html_url': repo['html_url'],
        'description': repo['description'],
        'fork': repo['fork'],
        'created_at': repo['created_at']
        
    } 
    return repo_fields
```

```python
def get_repos(token, since='512070020'):
    
    repos_details = []
    repos = list_repos(token)
    
    for repo in repos:
        try:
            owner = repo['owner']['login']
            name = repo['name']
            repo_details = get_repo_details(owner, name, token)
            repo_fields = extract_repo_fields(repo_details)
            repos_details.append(repo_fields)
        except KeyError:
            pass
        
    return repos_details
        
```

```python
repos_details = get_repos('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')
```

```python
def load_repos(repos_details, table):
    for repo in repos_details:
        table.put_item(Item=repo)
```
